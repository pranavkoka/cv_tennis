```python
from ultralytics import YOLO

model = YOLO('yolov8x')

result = model.predict("input\cv_image.png")
```

```python
print(result)
print("boxes:")
for box in result[0].boxes:
    print(box)
```

```python
result_vid = model.predict("input\cv_video_input.mp4")
```

```python
print(result_vid)
print("boxes:")
for box in result_vid[0].boxes:
    print(box)
```
