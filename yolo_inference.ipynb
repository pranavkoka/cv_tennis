```python
from ultralytics import YOLO

model = YOLO('yolov8x')

result = model.predict("input\cv_image.png", save=True, project=r'large_data')
```

```python
print(result)
print("boxes:")
for box in result[0].boxes:
    print(box)
```

```python
result_vid = model.predict("input\cv_video_input.mp4", save=True, project=r'large_data')
```

```python
print(result_vid)
print("boxes:")
for box in result_vid[0].boxes:
    print(box)
```
